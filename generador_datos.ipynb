Generar información con las siguientes columnas
- Nombre
- Edad
- Genero
- Ciudad
- Hijos
- Pesos
- Alturas

```python
import pandas as pd
from random import randint, choice
```

```python
lista_ciudades = ["Bogotá", "Chía", "Cajicá", "Zipa", "Tenjo", "Cogua"]
lista_generos = ["M", "F"]
nombres = []
edades = []
generos = []
ciudades = []
hijos = []
pesos = []
alturas = []
```

```python
for i in range(10000):
  nombres.append("persona_"+str(i))
  edades.append(randint(20, 70))
  generos.append(choice(lista_generos))
  ciudades.append(choice(lista_ciudades))
  hijos.append(randint(0, 3))
  pesos.append(randint(45, 100))
  alturas.append(randint(140, 190))
```

```python
data = {
    "nombres": nombres,
    "edades": edades,
    "generos": generos,
    "ciudades": ciudades,
    "hijos": hijos,
    "pesos": pesos,
    "alturas": alturas
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data = data, columns=["nombres", "edades", "generos", "ciudades", "hijos", "pesos", "alturas"])
```

```python
df_promedio_peso = df.groupby(["ciudades"])["pesos"].transform("mean")
df_promedio_peso
```

```python
df["mean_peso"] = df_promedio_peso
df.head(12)
```

```python
df_bajo_peso = df[df["pesos"] < df["mean_peso"]]
df_bajo_peso[["nombres", "ciudades", "edades", "pesos"]]
```

```python
df_edad_bajo_peso = df_bajo_peso[df_bajo_peso["edades"] > 69]
df_edad_bajo_peso
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
dfg = df_edad_bajo_peso.groupby(["ciudades"])["nombres"].count()
dfg
```

```python
dfg.plot.pie(
    title="Personas mayores por debajo del peso promedio"
)
```

```python
dfg_modified = df_edad_bajo_peso.groupby(["ciudades"])["nombres"].count().reset_index()
dfg_modified.columns = ["ciudades", "casos"]
dfg_modified["total"] = dfg_modified["casos"].sum()
dfg_modified["porcentaje"] = dfg_modified["casos"] / dfg_modified["total"] * 100
dfg_modified
```

```python
dfg_modified.index = dfg_modified["ciudades"]
dfg_modified.plot.pie(
    title="Personas mayores por debajo del peso promedio",
    y="porcentaje",
    autopct="%1.1f%%",
    legend=False,
)
```
